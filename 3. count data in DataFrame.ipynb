```python
import pandas as pd
```

```python
df = pd.read_csv("coursera_course_dataset_v3.csv")
df
```

```python
#count Distnict values - Descending
df["Organization"].value_counts()
```

```python
#count Distnict values - Ascending
df["Organization"].value_counts(ascending = True)
```

```python
# get Organization with out NaN
# by default the value_counts() functions returns all values wich are not null
# value_counts(dropna = False) to get null values to be counted.

df["Organization"].value_counts(dropna = False) 
```

```python
# get the Relative frequency - or percentage
df["Organization"].value_counts(normalize = True)
```

```python
#get specific item count : specific column
df["Organization"].value_counts()["IBM"]
```

```python
#count groupby unique values by Size : retuens in alphabetic order
df.groupby("Organization").size()
```

```python
#count groupby unique values by Count
df.groupby("Organization").count()
```

```python
#get a count of all colums across all colums
df.apply(pd.value_counts)
```
